# # Classificador de e-mail

---

## Dia 29 de outubro de 2025

---

Em 29 de outubro de 1971, foi enviado o primeiro e-mail da história, introduzindo o username@domainformato que ainda usamos. Hoje, existem bilhões de endereços de e-mail.

---

Neste desafio, você receberá uma lista de endereços de e-mail e precisará classificá-los alfabeticamente, primeiro pelo nome de domínio (a parte depois do " @") e, em seguida, pelo nome de usuário (a parte antes do "" @).

---

- A ordenação deve ignorar maiúsculas e minúsculas.
- Se mais de um e-mail tiver o mesmo domínio, classifique-os pelo nome de usuário.
- Retorna uma matriz com os endereços ordenados.
- Os remetentes devem manter a embalagem original.

Por exemplo, dado ["jill@mail.com", "john@example.com", "jane@example.com"], retorne ["jane@example.com", "john@example.com", "jill@mail.com"].

## Testes

- Esperando :1. sort(["jill@mail.com", "john@example.com", "jane@example.com"])deve retornar ["jane@example.com", "john@example.com", "jill@mail.com"].
- Esperando :2. sort(["bob@mail.com", "alice@zoo.com", "carol@mail.com"])deve retornar ["bob@mail.com", "carol@mail.com", "alice@zoo.com"].
- Esperando :3. sort(["user@z.com", "user@y.com", "user@x.com"])deve retornar ["user@x.com", "user@y.com", "user@z.com"].
- Esperando :4. sort(["sam@MAIL.com", "amy@mail.COM", "bob@Mail.com"])deve retornar ["amy@mail.COM", "bob@Mail.com", "sam@MAIL.com"].
- Esperando :5. sort(["simon@beta.com", "sammy@alpha.com", "Sarah@Alpha.com", "SAM@ALPHA.com", "Simone@Beta.com", "sara@alpha.com"])deve retornar ["SAM@ALPHA.com", "sammy@alpha.com", "sara@alpha.com", "Sarah@Alpha.com", "simon@beta.com", "Simone@Beta.com"].

````python
def sort(emails:str): return sorted( emails,  key=lambda item: (item.split("@")[1].lower(), item.split("@")[0].lower()))

sort(["jill@mail.com", "john@example.com", "jane@example.com"]) # deve retornar ["jane@example.com", "john@example.com", "jill@mail.com"].
sort(["bob@mail.com", "alice@zoo.com", "carol@mail.com"]) # ["bob@mail.com", "carol@mail.com", "alice@zoo.com"].
sort(["user@z.com", "user@y.com", "user@x.com"]) # ["user@x.com", "user@y.com", "user@z.com"].
sort(["sam@MAIL.com", "amy@mail.COM", "bob@Mail.com"]) # ["amy@mail.COM", "bob@Mail.com", "sam@MAIL.com"].
sort(["simon@beta.com", "sammy@alpha.com", "Sarah@Alpha.com", "SAM@ALPHA.com", "Simone@Beta.com", "sara@alpha.com"]) # ["SAM@ALPHA.com", "sammy@alpha.com", "sara@alpha.com", "Sarah@Alpha.com", "simon@beta.com", "Simone@Beta.com"].
````

# # Enésimo primo

---

## Dia 30 de outubro de 2025

---

Um número primo é um inteiro positivo maior que 1 que é divisível apenas por 1 e por ele mesmo. Os cinco primeiros números primos são 2, 3, 5, 7, e 11.

---

Dado um número inteiro positivo n, retorne o nn-ésimo número primo. Por exemplo, dado , 5retorne o 5º número primo: 11.

---

## Testes

1. nth_prime(5)deve retornar 11.
1. nth_prime(10)deve retornar 29.
1. nth_prime(16)deve retornar 53.
1. nth_prime(99)deve retornar 523.
1. nth_prime(1000)deve retornar 7919.

## Primeira Solução

````python
import time

start_time = time.time()

def nth_prime(n):
    count_number_prime:list = []
    count_divisor:int = 0
    number:int = 2

    while True:
        for i in range(1, number + 1):
            if number % i == 0:
                count_divisor += 1

        if count_divisor == 2:
            count_number_prime.append(number)

        count_divisor = 0
        number += 1

        if len(count_number_prime) == n:
            break

    return count_number_prime[-1]

# nth_prime(5)
# nth_prime(10)
# nth_prime(16)
# nth_prime(99)
nth_prime(1000)
end_time = time.time()
print(end_time - start_time)
````

## Segundo solução mais otimizada

````python
start_time = time.time()

def is_prime(number: int) -> bool:
    if number < 2:
        return False
    for i in range(2, int(number**0.5) + 1):
        if number % i == 0:
            return False
    return True

def nth_prime(n: int) -> int:
    primes: list = []
    number: int = 2

    while len(primes) < n:
        if is_prime(number):
            primes.append(number)
        number += 1
    return primes[-1]

nth_prime(1000)

end_time = time.time()
print(end_time - start_time)
````

# SpOoKy~Case

## Dia 31 de outubro de 2025

Dada uma string que representa o nome de uma variável, converta-a para "formato spooky case" usando as seguintes restrições:

- Substitua todos os sublinhados ( _) e hífens ( -) por um til ( ~).
- Coloque a primeira letra da sequência em maiúscula e, a partir daí, todas as outras letras em maiúscula, ignorando o caractere til (til) na contagem.

Por exemplo, dado hello_world, retorne HeLlO~wOrLd.

## Testes

1. Esperando :1. spookify("hello_world")deve retornar "HeLlO~wOrLd".
1. Esperando :2. spookify("Spooky_Case")deve retornar "SpOoKy~CaSe".
1. Esperando :3. spookify("TRICK-or-TREAT")deve retornar "TrIcK~oR~tReAt".
1. Esperando :4. spookify("c_a-n_d-y_-b-o_w_l")deve retornar "C~a~N~d~Y~~b~O~w~L".
1. Esperando :5. spookify("thE_hAUntEd-hOUsE-Is-fUll_Of_ghOsts")deve retornar "ThE~hAuNtEd~HoUsE~iS~fUlL~oF~gHoStS".

````python
def spookify(boo:str):
    lista:list = []
    txt:str = ""

    for val in boo:
        if not val in ["-", "_"]:
            txt += val
        else:
            lista.append(txt)
            txt = ""


    return lista

spookify("hello_world")                                                         # "HeLlO~wOrLd".
spookify("Spooky_Case")                                                         # "SpOoKy~CaSe".
# spookify("TRICK-or-TREAT")                                                      # "TrIcK~oR~tReAt".
# spookify("c_a-n_d-y_-b-o_w_l")                                                  # "C~a~N~d~Y~~b~O~w~L".
# spookify("thE_hAUntEd-hOUsE-Is-fUll_Of_ghOsts")                                 # "ThE~hAuNtEd~HoUsE~iS~fUlL~oF~gHoStS".
````

# Signature Validation

## Dia 01 de novembro de 2025

Given a message string, a secret key string, and a signature number, determine if the signature is valid using this encoding method:

- Letters in the message and secret key have these values:
    - a to z have values 1 to 26 respectively.
    - A to Z have values 27 to 52 respectively.
- All other characters have no value.
- Compute the signature by taking the sum of the message plus the sum of the secret key.

For example, given the message "foo" and the secret key "bar", the signature would be 57:

```
f (6) + o (15) + o (15) = 36
b (2) + a (1) + r (18) = 21
36 + 21 = 57
```

Check if the computed signature matches the provided signature.

## Tests

1. Failed:1. verify("foo", "bar", 57) should return True.
1. Failed:2. verify("foo", "bar", 54) should return False.
1. Failed:3. verify("freeCodeCamp", "Rocks", 238) should return True.
1. Failed:4. verify("Is this valid?", "No", 210) should return False.
1. Failed:5. verify("Is this valid?", "Yes", 233) should return True.
1. Failed:6. verify("Check out the freeCodeCamp podcast,", "in the mobile app", 514) should return True.


````python
import string

def verify(message, key, signature):

    MAPA_CARACTER_VALORES =
    {
        **dict(zip(string.ascii_lowercase, range(1, 27))),
        **dict(zip(string.ascii_uppercase, range(27, 53))),
        **dict(zip(string.punctuation, [0 for x in range(len(string.punctuation))])),
        **dict(zip(string.digits, [0 for x in range(len(string.digits))]))
    }

    return signature == sum(MAPA_CARACTER_VALORES.get(char, 0) for char in message) + sum(MAPA_CARACTER_VALORES.get(char, 0) for char in key)

verify("foo", "bar", 57)                                                        # should return True.
verify("foo", "bar", 54)                                                        # should return False.
verify("freeCodeCamp", "Rocks", 238)                                            # should return True.
verify("Is this valid?", "No", 210)                                             # should return False.
verify("Is this valid?", "Yes", 233)                                            # should return True.
verify("Check out the freeCodeCamp podcast,", "in the mobile app", 514)         # should return True.
````

# Infetado

## Dia 2 de novembro de 2025

Em 2 de novembro de 1988, foi lançado o primeiro grande worm de internet, infectando cerca de 10% dos computadores conectados à internet em apenas um dia.

Neste desafio, você recebe um número de dias que se passaram desde a liberação de um worm de internet e precisa determinar quantos computadores estão infectados usando as seguintes regras:

- No dia 0, o primeiro computador é infectado.
- A cada dia subsequente, o número de computadores infectados dobra.
- A cada 3 dias, após a disseminação do vírus, uma atualização é aplicada, reduzindo o número de computadores infectados em 20%. Arredonde o número de computadores atualizados para o número inteiro mais próximo.

## Por exemplo, em:

- Dia 0: 1 computador infectado no total.
- Dia 1: Um total de 2 computadores estão infectados.
- Dia 2: Um total de 4 computadores estão infectados.
- Dia 3: 8 computadores infectados no total. Em seguida, aplique a correção: 8 infectados * 20% = 1,6 corrigidos. Arredondando 1,6 para 2. 8 computadores infectados - 2 corrigidos = 6 computadores infectados no total após o dia 3.

Retorne o número total de computadores infectados após o período de dias especificado.

## Testes

1. Esperando :1. infected(1)deve retornar 2.
1. Esperando :2. infected(3)deve retornar 6.
1. Esperando :3. infected(8)deve retornar 152.
1. Esperando :4. infected(17)deve retornar 39808.
1. Esperando :5. infected(25)deve retornar 5217638.


````python
from math import ceil

def infected(days):
    cont:int = 1

    for x in range(days):
        cont = cont * 2
        if (x + 1) % 3 == 0:
            cont = cont - ceil(cont * 0.2)
    return cont

# infected(0)
# infected(1)
# infected(3)
# infected(8)
# infected(17)
infected(25)
````

## Dia 3 de novembro de 2025

- Palavras são quaisquer sequências de caracteres que não sejam espaços e são separadas por um único espaço.

# Pesquisa de imagens

## Dia 04 de novembro de 2025

Em 4 de novembro de 2001, o Google lançou sua busca de imagens, permitindo que as pessoas encontrassem imagens usando termos de pesquisa. Neste desafio, você irá simular a busca de imagens.

Dado um array de nomes de imagens e um termo de busca, retorne um array de nomes de imagens que contenham o termo de busca.
- Ignore as maiúsculas e minúsculas ao comparar os termos de pesquisa.
- Retorne as imagens na mesma ordem em que aparecem na matriz de entrada.

## Testes

1. Fracassado :1. image_search(["dog.png", "cat.jpg", "parrot.jpeg"], "dog")deve retornar ["dog.png"].
1. Fracassado :2. image_search(["Sunset.jpg", "Beach.png", "sunflower.jpeg"], "sun")deve retornar ["Sunset.jpg", "sunflower.jpeg"].
1. Fracassado :3. image_search(["Moon.png", "sun.jpeg", "stars.png"], "PNG")deve retornar ["Moon.png", "stars.png"].
1. Fracassado :4. image_search(["cat.jpg", "dogToy.jpeg", "kitty-cat.png", "catNip.jpeg", "franken_cat.gif"], "Cat")deve retornar ["cat.jpg", "kitty-cat.png", "catNip.jpeg", "franken_cat.gif"].

````python
def image_search(images, term): return [val for val in images if term.lower() in val.lower()]

image_search(["dog.png", "cat.jpg", "parrot.jpeg"], "dog")                                           # deve retornar ["dog.png"].
image_search(["Sunset.jpg", "Beach.png", "sunflower.jpeg"], "sun")                                   # deve retornar ["Sunset.jpg", "sunflower.jpeg"].
image_search(["Moon.png", "sun.jpeg", "stars.png"], "PNG")                                           # deve retornar ["Moon.png", "stars.png"].
image_search(["cat.jpg", "dogToy.jpeg", "kitty-cat.png", "catNip.jpeg", "franken_cat.gif"], "Cat")   # deve retornar ["cat.jpg", "kitty-cat.png", "catNip.jpeg", "franken_cat.gif"]
````
