# # Classificador de e-mail

---

## Dia 29 de outubro de 2025

---

Em 29 de outubro de 1971, foi enviado o primeiro e-mail da história, introduzindo o username@domainformato que ainda usamos. Hoje, existem bilhões de endereços de e-mail.

---

Neste desafio, você receberá uma lista de endereços de e-mail e precisará classificá-los alfabeticamente, primeiro pelo nome de domínio (a parte depois do " @") e, em seguida, pelo nome de usuário (a parte antes do "" @).

---

- A ordenação deve ignorar maiúsculas e minúsculas.
- Se mais de um e-mail tiver o mesmo domínio, classifique-os pelo nome de usuário.
- Retorna uma matriz com os endereços ordenados.
- Os remetentes devem manter a embalagem original.

Por exemplo, dado ["jill@mail.com", "john@example.com", "jane@example.com"], retorne ["jane@example.com", "john@example.com", "jill@mail.com"].

## Testes

- Esperando :1. sort(["jill@mail.com", "john@example.com", "jane@example.com"])deve retornar ["jane@example.com", "john@example.com", "jill@mail.com"].
- Esperando :2. sort(["bob@mail.com", "alice@zoo.com", "carol@mail.com"])deve retornar ["bob@mail.com", "carol@mail.com", "alice@zoo.com"].
- Esperando :3. sort(["user@z.com", "user@y.com", "user@x.com"])deve retornar ["user@x.com", "user@y.com", "user@z.com"].
- Esperando :4. sort(["sam@MAIL.com", "amy@mail.COM", "bob@Mail.com"])deve retornar ["amy@mail.COM", "bob@Mail.com", "sam@MAIL.com"].
- Esperando :5. sort(["simon@beta.com", "sammy@alpha.com", "Sarah@Alpha.com", "SAM@ALPHA.com", "Simone@Beta.com", "sara@alpha.com"])deve retornar ["SAM@ALPHA.com", "sammy@alpha.com", "sara@alpha.com", "Sarah@Alpha.com", "simon@beta.com", "Simone@Beta.com"].

```python
def sort(emails:str): return sorted( emails,  key=lambda item: (item.split("@")[1].lower(), item.split("@")[0].lower()))

sort(["jill@mail.com", "john@example.com", "jane@example.com"]) # deve retornar ["jane@example.com", "john@example.com", "jill@mail.com"].
sort(["bob@mail.com", "alice@zoo.com", "carol@mail.com"]) # ["bob@mail.com", "carol@mail.com", "alice@zoo.com"].
sort(["user@z.com", "user@y.com", "user@x.com"]) # ["user@x.com", "user@y.com", "user@z.com"].
sort(["sam@MAIL.com", "amy@mail.COM", "bob@Mail.com"]) # ["amy@mail.COM", "bob@Mail.com", "sam@MAIL.com"].
sort(["simon@beta.com", "sammy@alpha.com", "Sarah@Alpha.com", "SAM@ALPHA.com", "Simone@Beta.com", "sara@alpha.com"]) # ["SAM@ALPHA.com", "sammy@alpha.com", "sara@alpha.com", "Sarah@Alpha.com", "simon@beta.com", "Simone@Beta.com"].
```

# # Enésimo primo

---

## Dia 30 de outubro de 2025

---

Um número primo é um inteiro positivo maior que 1 que é divisível apenas por 1 e por ele mesmo. Os cinco primeiros números primos são 2, 3, 5, 7, e 11.

---

Dado um número inteiro positivo n, retorne o nn-ésimo número primo. Por exemplo, dado , 5retorne o 5º número primo: 11.

---

## Testes

1. nth_prime(5)deve retornar 11.
1. nth_prime(10)deve retornar 29.
1. nth_prime(16)deve retornar 53.
1. nth_prime(99)deve retornar 523.
1. nth_prime(1000)deve retornar 7919.

## Primeira Solução

```python
import time

start_time = time.time()

def nth_prime(n):
    count_number_prime:list = []
    count_divisor:int = 0
    number:int = 2

    while True:
        for i in range(1, number + 1):
            if number % i == 0:
                count_divisor += 1

        if count_divisor == 2:
            count_number_prime.append(number)

        count_divisor = 0
        number += 1

        if len(count_number_prime) == n:
            break

    return count_number_prime[-1]

# nth_prime(5)
# nth_prime(10)
# nth_prime(16)
# nth_prime(99)
nth_prime(1000)
end_time = time.time()
print(end_time - start_time)
```

## Segundo solução mais otimizada

```python
start_time = time.time()

def is_prime(number: int) -> bool:
    if number < 2:
        return False
    for i in range(2, int(number**0.5) + 1):
        if number % i == 0:
            return False
    return True

def nth_prime(n: int) -> int:
    primes: list = []
    number: int = 2

    while len(primes) < n:
        if is_prime(number):
            primes.append(number)
        number += 1
    return primes[-1]

nth_prime(1000)

end_time = time.time()
print(end_time - start_time)
```
